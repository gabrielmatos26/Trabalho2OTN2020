```python
resultados_modelo_250_max_jogadas_40_rodadas_10 = pd.read_csv(
    "resultados/resultados_modelo_250_max_jogadas_40_rodadas_10.csv", index_col=0)

resultados_modelo_500_max_jogadas_40_rodadas_10 = pd.read_csv(
    "resultados/resultados_modelo_500_max_jogadas_40_rodadas_10.csv", index_col=0)

resultados_modelo_250_max_jogadas_300_rodadas_10 = pd.read_csv(
    "resultados/resultados_modelo_250_max_jogadas_300_rodadas_10.csv", index_col=0)

resultados_modelo_500_max_jogadas_300_rodadas_10 = pd.read_csv(
    "resultados/resultados_modelo_500_max_jogadas_300_rodadas_10.csv", index_col=0)
```

```python
medias_modelo_500_max_jogadas_300_rodadas_10 = resultados_modelo_500_max_jogadas_300_rodadas_10[['Modelo', 'Tempo de treino (em horas)']].copy()
medias_modelo_500_max_jogadas_300_rodadas_10['SR médio peças brancas'] = resultados_modelo_500_max_jogadas_300_rodadas_10.values[:,2:12].mean(axis=1)
medias_modelo_500_max_jogadas_300_rodadas_10['SR médio peças pretas'] = resultados_modelo_500_max_jogadas_300_rodadas_10.values[:,12:22].mean(axis=1)
medias_modelo_500_max_jogadas_300_rodadas_10['Taxa média de empates'] = resultados_modelo_500_max_jogadas_300_rodadas_10.values[:,22:].mean(axis=1)
print("10 rodadas de 500 jogos do modelo com 500 Gerações e com Máximo de 300 jogadas sem capturar peças")
medias_modelo_500_max_jogadas_300_rodadas_10
```

```python
single_sigma_aleatório_brancas = np.array([resultados_modelo_250_max_jogadas_40_rodadas_10.values[0,2:12],
                                          resultados_modelo_500_max_jogadas_40_rodadas_10.values[0,2:12],
                                          resultados_modelo_250_max_jogadas_300_rodadas_10.values[0,2:12],
                                          resultados_modelo_500_max_jogadas_300_rodadas_10.values[0,2:12]])

single_sigma_aleatório_empates = np.array([resultados_modelo_250_max_jogadas_40_rodadas_10.values[0,22:],
                                          resultados_modelo_500_max_jogadas_40_rodadas_10.values[0,22:],
                                          resultados_modelo_250_max_jogadas_300_rodadas_10.values[0,22:],
                                          resultados_modelo_500_max_jogadas_300_rodadas_10.values[0,22:]])

n_sigma_aleatório_brancas = np.array([resultados_modelo_250_max_jogadas_40_rodadas_10.values[1,2:12],
                                          resultados_modelo_500_max_jogadas_40_rodadas_10.values[1,2:12],
                                          resultados_modelo_250_max_jogadas_300_rodadas_10.values[1,2:12],
                                          resultados_modelo_500_max_jogadas_300_rodadas_10.values[1,2:12]])

n_sigma_aleatório_empates = np.array([resultados_modelo_250_max_jogadas_40_rodadas_10.values[1,22:],
                                          resultados_modelo_500_max_jogadas_40_rodadas_10.values[1,22:],
                                          resultados_modelo_250_max_jogadas_300_rodadas_10.values[1,22:],
                                          resultados_modelo_500_max_jogadas_300_rodadas_10.values[1,22:]])

aleatório_brancas_single_sigma = np.array([resultados_modelo_250_max_jogadas_40_rodadas_10.values[2,12:22],
                                          resultados_modelo_500_max_jogadas_40_rodadas_10.values[2,12:22],
                                          resultados_modelo_250_max_jogadas_300_rodadas_10.values[2,12:22],
                                          resultados_modelo_500_max_jogadas_300_rodadas_10.values[2,12:22]])

aleatório_single_sigma_empates = np.array([resultados_modelo_250_max_jogadas_40_rodadas_10.values[2,22:],
                                          resultados_modelo_500_max_jogadas_40_rodadas_10.values[2,22:],
                                          resultados_modelo_250_max_jogadas_300_rodadas_10.values[2,22:],
                                          resultados_modelo_500_max_jogadas_300_rodadas_10.values[2,22:]])

aleatório_brancas_n_sigma = np.array([resultados_modelo_250_max_jogadas_40_rodadas_10.values[3,12:22],
                                          resultados_modelo_500_max_jogadas_40_rodadas_10.values[3,12:22],
                                          resultados_modelo_250_max_jogadas_300_rodadas_10.values[3,12:22],
                                          resultados_modelo_500_max_jogadas_300_rodadas_10.values[3,12:22]])

aleatório_n_sigma_empates = np.array([resultados_modelo_250_max_jogadas_40_rodadas_10.values[3,22:],
                                          resultados_modelo_500_max_jogadas_40_rodadas_10.values[3,22:],
                                          resultados_modelo_250_max_jogadas_300_rodadas_10.values[3,22:],
                                          resultados_modelo_500_max_jogadas_300_rodadas_10.values[3,22:]])
```

```python
try:
    if fig is not None:
        plt.close(fig)
except:
    pass
fig, ax = plt.subplots(1, 2, figsize=(10,5))
ax[0].set_title("Taxa de vitórias do Modelo com 1 sigma(peças brancas) contra\n"+
                "Modelo aleatório(peças pretas) após 10 execuções de 500 jogos\n",
            fontsize=9)
ax[1].set_title("Taxa de vitórias do Modelo com N sigmas(peças brancas) contra\n"+
                "Modelo aleatório(peças pretas) após 10 execuções de 500 jogos\n",
            fontsize=9)

bplot1 = ax[0].boxplot(single_sigma_aleatório_brancas.T, patch_artist=True)
bplot2 = ax[1].boxplot(n_sigma_aleatório_brancas.T, patch_artist=True)

ax[0].set_yticks(np.arange(0.32, 0.71, 0.05))
ax[0].set_ylabel("Taxa de sucesso")
ax[1].set_yticks(np.arange(0.32, 0.71, 0.05))

bplots = [bplot1, bplot2]
colors = ['pink', 'lightblue', 'lightgreen', 'lightyellow']
for bplot in bplots:
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

textstr = '\n'.join((
    "Legenda eixo x:",
    "1 - 250 Gerações; Máximo de 40 jogadas sem capturar peças",
    "2 - 500 Gerações; Máximo de 40 jogadas sem capturar peças",
    "3 - 250 Gerações; Máximo de 300 jogadas sem capturar peças",
    "4 - 500 Gerações; Máximo de 300 jogadas sem capturar peças"))
# these are matplotlib.patch.Patch properties
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)

# place a text box in upper left in axes coords
plt.text(0.6, -0.15, textstr, transform=ax[0].transAxes, fontsize=8,
        verticalalignment='top', bbox=props)

plt.subplots_adjust(bottom=0.25)
plt.show()
```
